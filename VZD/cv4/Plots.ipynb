```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
iris_log = pd.read_csv("Iris.log", sep="\t")
sonar_log = pd.read_csv("Sonar.log", sep="\t")
weighting_log = pd.read_csv("Weighting.log", sep="\t")
```

```python
dfs = {'Iris': iris_log, 
       'Sonar': sonar_log,
       'Weighting': weighting_log}
```

```python
for df_name, df in dfs.items():
    print('*'*50)
    print("{df}".format(df=df_name))
    for key, group in df.groupby("Distance"):
        print(key, '\t', "Max Perf:", str(group.Performance.max()), "\twith k =", str(group.k.get(group.Performance.idxmax)))
        plt.plot(group.k, group.Performance, label=key)

    plt.legend(loc='best')
    plt.show()
```
